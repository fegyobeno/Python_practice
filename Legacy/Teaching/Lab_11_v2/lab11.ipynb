```python
from keras.models import Sequential
from keras.layers import Dense, Input
from keras.optimizers import Adam
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
np.random.rand()
```

```python
x_0 = np.array([x for x in range(100)])
y_0 = np.array([x for x in range(100)])
```

```python
x_0 = x_0.reshape(-1, 1)
y_0 = y_0.reshape(-1, 1)
```

```python
model1 = Sequential() # a rétegek sorban jönnek, lineárisan
model1.add(Input(shape=(1,))) # A bemeneti rétegnek egy feature-je van (az y érték)
model1.add(Dense(1, activation='linear')) # A rejtett rétegnek két neuront adunk, ezeket fogja tudni optimalizálni w+b
model1.add(Dense(1)) # A kimenet lineáris -> megfelelő, mivel folytonos értékeket szeretnénk jósolni

model1.compile(optimizer=Adam(learning_rate=0.01), loss='mse', metrics=['mae'])
model1.summary()
```

```python
model1.fit(x_0, y_0, epochs=100, verbose=0, validation_split=0.2)
```

```python
plt.scatter(x_0,y_0)
y_pred = model1.predict(x_0)
plt.plot(x_0, y_pred, color='red')
plt.show()
```

```python
x = np.array([x for x in range(100)])
y = np.array([10 + x + (np.random.rand()*10*np.random.choice([-1,1])) for x in range(100)])
x = x.reshape(-1,1)
y = y.reshape(-1,1)
```

```python
print(f"{x.shape} {y.shape}")
```

```python
plt.scatter(x,y)
```

```python
model1 = Sequential() # a rétegek sorban jönnek, lineárisan
model1.add(Input(shape=(1,))) # A bemeneti rétegnek egy feature-je van (az y érték)
model1.add(Dense(1, activation='linear')) # A rejtett rétegnek két neuront adunk, ezeket fogja tudni optimalizálni w+b
model1.add(Dense(1)) # A kimenet lineáris -> megfelelő, mivel folytonos értékeket szeretnénk jósolni

model1.compile(optimizer=Adam(learning_rate=0.01), loss='mse', metrics=['mae'])
model1.summary()
```

```python
model2 = Sequential() # a rétegek sorban jönnek, lineárisan
model2.add(Input(shape=(1,))) # A bemeneti rétegnek egy feature-je van (az y érték)
model2.add(Dense(2, activation='linear')) # A rejtett rétegnek két neuront adunk, ezeket fogja tudni optimalizálni w+b
model2.add(Dense(1)) # A kimenet lineáris -> megfelelő, mivel folytonos értékeket szeretnénk jósolni

model2.compile(optimizer=Adam(learning_rate=0.01), loss='mse', metrics=['mae'])
model2.summary()
```

```python
model1.fit(x, y, epochs=100, verbose=0, validation_split=0.2)
model2.fit(x, y, epochs=100, verbose=0, validation_split=0.2)
```

```python
plt.scatter(x,y)
y_pred = model1.predict(x)
plt.plot(x, y_pred, color='red')
plt.show()
```

```python
plt.scatter(x,y)
y_pred = model2.predict(x)
plt.plot(x, y_pred, color='red')
plt.show()
```

```python
y_pred_model1 = model1.predict(x)
_, accuracy_model1 = model1.evaluate(x, y, verbose=0) # Using model's built in evaluation
print(f"Model 1 Accuracy: {accuracy_model1}")

y_pred_model2 = model2.predict(x)
_, accuracy_model2 = model2.evaluate(x, y, verbose=0) # Using model's built in evaluation

print(f"Model 2 Accuracy: {accuracy_model2}")
```

```python
import pandas as pd

url = "https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv"
data = pd.read_csv(url)
data.head()
```

```python
import seaborn as sns

correlation_matrix = data.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Boston Housing Dataset')
plt.show()
```

```python
X = data[['rm', 'lstat']]  # average rooms and percentage of lower status of the population
y = data['medv']         # median house value

print(X.head())
print(y.head())
```

```python
from mpl_toolkits.mplot3d import Axes3D  # Needed for 3D plotting

np.random.seed(0)
X1 = data['rm']
X2 = data['age']
Y = data['medv']


# Create 3D scatter plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X1, X2, Y, c='blue', marker='o', alpha=0.7)

ax.set_xlabel('Room Number')
ax.set_ylabel('Age of the house')
ax.set_zlabel('Median Value')
ax.set_title('3D Scatter Plot of Dataset (Two Inputs, One Output)')

plt.show()
```

```python
print(f"{X.shape} {Y.shape}")
```

```python
model_boston = Sequential()
model_boston.add(Input(shape=(2,)))
model_boston.add(Dense(16, activation='relu'))
model_boston.add(Dense(8, activation='relu'))
model_boston.add(Dense(1))

model_boston.compile(optimizer=Adam(learning_rate=0.01), loss='mse', metrics=['mae'])
model_boston.summary()
```

```python
model_boston.fit(X, Y, epochs=100, verbose=0, validation_split=0.2)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # Needed for 3D plotting

# Assuming 'data' is your DataFrame with columns 'rm', 'age', 'medv'
X1 = data['rm'].values
X2 = data['age'].values
Y = data['medv'].values

# Prepare input for prediction: shape (n_samples, 2)
X_input = np.column_stack((X1, X2))

# Get model predictions
Y_pred = model_boston.predict(X_input).flatten()

# Create 3D scatter plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot original data points
ax.scatter(X1, X2, Y, c='blue', marker='o', alpha=0.7, label='True Values')

# Plot predicted points
ax.scatter(X1, X2, Y_pred, c='red', marker='^', alpha=0.7, label='Predicted Values')

ax.set_xlabel('Room Number')
ax.set_ylabel('Age of the house')
ax.set_zlabel('Median Value')
ax.set_title('3D Scatter Plot of Dataset with Model Predictions')

ax.legend()

plt.show()
```

```python
#Lets check the accuracy of the model

y_pred = model_boston.predict(X)
err = 0
for i in range(len(y_pred)):
    err += abs(y_pred[i]-y[i])

print(f"The average house price is {(sum(Y)/len(Y))*1000}$")
print(f"The error on average is {1000*(err/len(y_pred))}$")
print(f"The error is {100*((err/len(y_pred))/(sum(Y)/len(Y)))}%")
```
